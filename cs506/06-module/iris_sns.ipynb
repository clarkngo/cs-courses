```python
# seaborn is a data-viz lib as an interface to matplotlib along with highlevel (aka fancy) plot capabilities 
# this works with pandas dataframe.
import seaborn as sns
import pandas as pd
# defining a dataframe
iris = pd.read_csv('C:\\CityU\\CS506\\Data\\iris.data.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
print(iris.head())
iris['petal_width']

# sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
# if you like to modify the labels for axis, you need the axis object to 
# change
ax = sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
ax.set(xlabel = "sepal_length (cm)", ylabel = "sepal_width (cm)")
```

```python
# let's look at the plotting per class using seborn
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='class')

```

```python
# how about the line chart?
# still a simple format
sns.lineplot(data=iris.drop(['class'], axis=1))
```

```python
# Here is the histogram
wine_reviews = pd.read_csv('C:\\CityU\\CS506\\Data\\winemag-data-130k-v2.csv', index_col=0) #index_col chooses the row label if already provided
#index_col = False when to ignore the first col. as the index, or useful when EOL has a bad character(s) to be skipped

from scipy.stats import norm
sns.distplot(wine_reviews['points'], bins=20, kde=False, color='green', fit=norm, vertical=True) #kde = Whether to plot a gaussian kernel density estimate.

```

```python
# bar chart now

sns.countplot(wine_reviews['points'])
```

```python
# Heatmap visualizes correlation between 2 features in a matrix form as colors.
ax = sns.heatmap(iris.corr(), center=True, annot=True)

ax.set_ylim(0, 4)
ax.set_xlim(0, 4)
```

```python
# Faceting is technique that separates each individual feature as a set of subplots in a single figure
# in Seaborn, we have FacetGrid method. It also works with categorical data set.
# - pass either row or column depending on how you like to slice
# - use a map method with a plot type to that FacetGrid object and the set of features (column data)

g = sns.FacetGrid(iris, col = 'class')
g = g.map(sns.kdeplot, 'sepal_length')
```

```python
# Pairplot
# shows a grid of pair-wise plot in a dataset
# diagonal shows the histogram of the feature (col.data)
# otherwise, each figure shows x, y in two dimensional coordinate between 2 features.
# this is useful to have the overview of the value distribution.
sns.pairplot(iris, hue='class')
```
