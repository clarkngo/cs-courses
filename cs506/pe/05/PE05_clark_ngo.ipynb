## Load the image

### Subtask:
Load the `PE05-input.jpg` image into a Pillow Image object.


**Reasoning**:
Load the image from the file into a Pillow Image object.



```python
from PIL import Image

image = Image.open('/content/PE05-input.jpg')
```

## Apply filters

### Subtask:
Experiment with different image filters from `PIL.ImageFilter` and display the results.


**Reasoning**:
Apply different filters from `PIL.ImageFilter` to the loaded image and display the results for each filter.



```python
from PIL import ImageFilter

# Apply BLUR filter
blurred_image = image.filter(ImageFilter.BLUR)
print("BLUR filter applied:")
display(blurred_image)

# Apply CONTOUR filter
contour_image = image.filter(ImageFilter.CONTOUR)
print("CONTOUR filter applied:")
display(contour_image)

# Apply SHARPEN filter
sharpened_image = image.filter(ImageFilter.SHARPEN)
print("SHARPEN filter applied:")
display(sharpened_image)
```

## Apply enhancers

### Subtask:
Experiment with different image enhancers from `PIL.ImageEnhance` and display the results.


**Reasoning**:
Import the ImageEnhance module and apply different image enhancers as instructed.



```python
from PIL import ImageEnhance

# Enhance color
color_enhancer = ImageEnhance.Color(image)
enhanced_color_image = color_enhancer.enhance(1.5)
print("Color enhancement (factor 1.5):")
display(enhanced_color_image)

# Enhance brightness
brightness_enhancer = ImageEnhance.Brightness(image)
enhanced_brightness_image = brightness_enhancer.enhance(1.5)
print("Brightness enhancement (factor 1.5):")
display(enhanced_brightness_image)

# Enhance contrast
contrast_enhancer = ImageEnhance.Contrast(image)
enhanced_contrast_image = contrast_enhancer.enhance(1.5)
print("Contrast enhancement (factor 1.5):")
display(enhanced_contrast_image)

# Enhance sharpness
sharpness_enhancer = ImageEnhance.Sharpness(image)
enhanced_sharpness_image = sharpness_enhancer.enhance(1.5)
print("Sharpness enhancement (factor 1.5):")
display(enhanced_sharpness_image)
```
