```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data file
file_name = 'PE04_winereview.csv'
# Assuming the CSV file is available in the execution environment
df = pd.read_csv(file_name)

print("--- Data Visualization Tasks for PE04 ---")

# ----------------------------------------------------------------------
# Part 1: Bar Graph (Value Counts) for 'points'
# Uses value_counts() to count the frequency of each score.
print("\n--- Part 1: Bar Graph (Value Counts) for 'points' ---")

# Calculate value counts and sort by index (point value)
points_counts = df['points'].value_counts().sort_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
points_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Wine Review Scores (Points)')
plt.xlabel('Score (Points)')
plt.ylabel('Frequency (Number of Wines)')
plt.grid(axis='y', alpha=0.5)
plt.xticks(rotation=0)
plt.show()
```

```python

# ----------------------------------------------------------------------
# Part 2: Histograms for 'points' (Bins 10 and 3)
# Uses the hist() method with specified binning logic.
print("\n--- Part 2: Histograms for 'points' (Bins 10 and 3) ---")

min_point, max_point = df['points'].min(), df['points'].max() # Points range from 80 to 100.

# Bin Size 10: 10 equal-width bins across the range [80, 100]
bins_10_points = np.linspace(min_point, max_point, 11)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
df['points'].hist(bins=bins_10_points, edgecolor='black', linewidth=1.5, alpha=0.7)
plt.title('Points Distribution (10 Equal Bins)')
plt.xlabel('Score (Points)')
plt.ylabel('Frequency')
plt.xticks(bins_10_points)
plt.grid(axis='y', alpha=0.5)

# Bin Size 3 (Custom width bins: [80-82], [83-85], [86-88], etc.)
bins_3_points_custom = np.arange(min_point, max_point + 3, 3)

plt.subplot(1, 2, 2)
df['points'].hist(bins=bins_3_points_custom, edgecolor='black', linewidth=1.5, alpha=0.7)
plt.title('Points Distribution (Bin Width 3)')
plt.xlabel('Score Range (Points)')
plt.ylabel('Frequency')
# Setting custom labels for clarity
plt.xticks(bins_3_points_custom[:-1], [f'{i}-{i+2}' for i in range(min_point, max_point + 1, 3)], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


# ----------------------------------------------------------------------
# Part 3: Histograms for 'price' (Bins 10 and 3)
# Requires handling missing values (NaNs) and considering the skewed data distribution.
print("\n--- Part 3: Histograms for 'price' (Bins 10 and 3) ---")
price_data = df['price'].dropna()
max_price = price_data.max()
bins_10_price = 10 
price_cap = 100
price_data_capped = price_data[price_data <= price_cap]
bins_3_price_custom = np.arange(0, price_cap + 3, 3)

plt.figure(figsize=(12, 5))

# Plot 1: Bin Size 10 (Full Range)
plt.subplot(1, 2, 1)
price_data.hist(bins=bins_10_price, edgecolor='black', linewidth=1.5, alpha=0.7)
plt.title('Price Distribution (10 Equal Bins - Full Range)')
# Displaying bin size in the label
plt.xlabel(f'Price (Binned by ${max_price/bins_10_price:.0f}s)') 
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)

# Plot 2: Bin Width 3 (Capped at $100)
plt.subplot(1, 2, 2)
price_data_capped.hist(bins=bins_3_price_custom, edgecolor='black', linewidth=1.5, alpha=0.7)
# Using unicode character () to avoid LaTeX parsing error
plt.title(f'Price Distribution (Bin Width $3 - Price  ${price_cap})', fontsize=12)
plt.xlabel('Price ($)', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.xticks(np.arange(0, price_cap + 1, 10))
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
