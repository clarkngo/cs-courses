```python
# if you entered \n in input method, it wont treat it as new line character rather two seperate characters.
# Here is how you can handle.
delimiter = "#NEWLINE#"
print(f"Enter your text, use {delimiter} to mark a newline:")
text = input()
lines = text.split(delimiter)

# Process the lines as needed
print("You entered:", lines)
```

```python
# With escape sequence processing
text = input("Enter text: ")
print(text)  # If the user enters "Hello\nWorld", it will print on two lines

# Without escape sequence processing (raw string)
text = input(r"Enter text: ")
print(text)  # If the user enters "Hello\nWorld", it will print "Hello\nWorld"
```

```python
#
# -- traditionally, we have been using open -- close a file for any operation
#
f = open('test.txt', 'r')
contents = f.read()
print (contents)
f.close()
```

```python
#
# -- new way of opening a file with the context manager
#
with open('test.txt', 'r') as f:
    contents = f.read()
    print (contents)
    # no need to close with the context manager!!
```

# Logging

```python
import logging
import os

# Properly remove the logger
def remove_logger(logger):
    # Remove all handlers associated with the logger
    for handler in logger.handlers[:]:  # Copy the handlers list
        logger.removeHandler(handler)  # Remove the handler from the logger
        handler.close()  # Close the handler to release resources
        del handler  # Delete the handler to remove it from memory
# Create a custom logger with some name. may be cs506
logger = logging.getLogger('cs506')
logger.propagate = False
# Validating if logger had any handlers. If so, printing their names
for handler in logger.handlers[:]:  # Copy the handlers list
    print(handler.name)
    print(handler.level)

if logger.hasHandlers():
  print("Logger has handlers")
else:
  print("Logger has no handlers")


```

```python
from pickle import TRUE
if not logger.hasHandlers():
  print("Enetered log handler section")
  # Set the logging level (DEBUG, INFO, WARNING, ERROR, CRITICAL)
  # Create handlers
  console_handler = logging.StreamHandler()  # Logs to console
  console_handler.setLevel(logging.DEBUG)
  console_handler.set_name('console_handler')

  file_handler = logging.FileHandler('app.log')  # Logs to a file
  file_handler.set_name('file_handler')
  file_handler.setLevel(logging.ERROR)  # Only log errors and above to the file

  # Create formatters and add them to the handlers
  console_format = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
  file_format = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

  console_handler.setFormatter(console_format)
  file_handler.setFormatter(file_format)

  # Add handlers to the logger
  logger.addHandler(console_handler)
  logger.addHandler(file_handler)



# Example log messages
def divide(a, b):
    try:
        logger.info("Divide function called with %s and %s", a, b)
        result = a / b
        logger.debug("Division result: %s", result)
        return result
    except ZeroDivisionError as e:
        logger.error("Error occurred: %s", e, exc_info=TRUE)
        return None

def main():
    logger.info("Application started")
    divide(10, 2)
    divide(5, 0)
    logger.info("Application finished")



if __name__ == "__main__":
  logger.setLevel(logging.DEBUG)
  main()
  # Remove the logger
  remove_logger(logger)
```
