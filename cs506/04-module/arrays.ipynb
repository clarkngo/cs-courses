```python
import numpy as np
```

```python
arr = np.array([1, 2, 3])
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr[0], arr[1], arr[2])
arr[2] =4
print(arr)

arr2= np.array([[1, 2, 3], [4, 5, 6]])
print(arr2.shape)
print(arr2[0, 0], arr2[0, 1], arr2[1, 0])
print(arr2[-1])
```

```python
zero = np.zeros((2, 2))
print(zero)

ones = np.ones((1, 2))
print(ones)

constant = np.full((2, 2), 5)
print(constant)

diag=np.eye(3)
print(diag) 

diag = np.eye(4, k=1)
print(diag)

rand = np.random.random((2, 2))
print(rand)
```
