```python
def first(n):
    return n[0]

def sort_list_first(tuples):
    return sorted(tuples, key=first)

print(sort_list_first([(5, 2), (2, 1), (4, 4), (3, 2), (1, 2)]))
```

```python
# List of students
students = ["Alice", "Bob", "Charlie", "Alice", "Diana", "Bob"]

# Remove duplicates using a set
unique_students = set(students)
print("Unique Students:", unique_students)

# Create a dictionary with student and their club
club_membership = {
    "Alice": "Art Club",
    "Bob": "Science Club",
    "Charlie": "Drama Club",
    "Diana": "Math Club"
}

# Accessing membership
for student in unique_students:
    if student in club_membership:
        print(f"{student} is in {club_membership[student]}")
    else:
        print(f"{student} is not in any club")

# Add a new student and assign them to a club
club_membership["Eve"] = "Music Club"
print("Updated Club Membership:", club_membership)

# Remove a student from the list
del club_membership["Bob"]
print("After Bob left:", club_membership)
```

```python
# Haunted Library Inventory System üëªüìö

# Step 1: Raw list of haunted items (some are duplicates)
raw_items = ["Grimoire of Shadows", "Cursed Mirror", "Grimoire of Shadows",
             "Phantom Candle", "Cursed Mirror", "Blood-Stained Map"]

print("üì¶ Items found scattered around the library...")
print(raw_items)

# Step 2: Use a set to remove duplicates
unique_items = set(raw_items)
print("\nüßπ After removing duplicates, unique haunted items are:")
print(unique_items)

# Step 3: Assign item types using a dictionary
item_catalog = {
    "Grimoire of Shadows": "Ancient Spellbook",
    "Cursed Mirror": "Enchanted Object",
    "Phantom Candle": "Ghostly Artifact",
    "Blood-Stained Map": "Lost Relic"
}

# Step 4: Access and display each item's type
print("\nüìñ Cataloged Items:")
for item in unique_items:
    print(f"- {item}: {item_catalog[item]}")

# Step 5: Add a new cursed object
print("\n‚ö†Ô∏è A new haunted item has appeared: 'Whispering Skull'!")
item_catalog["Whispering Skull"] = "Cursed Skull"
unique_items.add("Whispering Skull")

# Display updated catalog
print("\nüîÅ Updated Item Catalog:")
for item in unique_items:
    print(f"- {item}: {item_catalog[item]}")

# Step 6: Remove a destroyed item
print("\nüî• 'Phantom Candle' has been destroyed in a ritual.")
unique_items.remove("Phantom Candle")
del item_catalog["Phantom Candle"]

# Final state
print("\nüßæ Final Inventory Report:")
for item in unique_items:
    print(f"- {item}: {item_catalog[item]}")
```
