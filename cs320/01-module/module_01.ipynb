```python
import random
num = random.randint(20, 25)
randomNumber = None

print("Try to guess your number: ")
print("--------------------------------------------------------")
while randomNumber != num:

    numRan = input("Pick a number between 20 and 25: ")
    guess = int(numRan)

    if guess == num:
        print("Perfect, you did it!", guess)
        break
    else:
        print("Sorry. Try again please!", guess)
```
